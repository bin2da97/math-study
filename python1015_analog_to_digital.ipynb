```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# 아날로그 신호 생성 (sine wave)
t = np.linspace(0, 2 * np.pi, 1000) #간격이 있지만 ,1000개이므로 연속인것 보인다 
analog_signal = np.sin(t)

# digital 신호를 생성하기 위한 sampling
sample_points_x = np.linspace(0, 2 * np.pi, 30)
sample_points_y = np.sin(sample_points_x)

# Create a figure and axis
fig, ax = plt.subplots()
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1, 1)

# Create lines for the analog signal and digital samples
analog_line, = ax.plot([], [], lw=2, label="Analog Signal", color='blue')
digital_samples, = ax.plot([], [], 'ro', markersize=8, label="Digital Samples")


# Add labels and legend
ax.set_xlabel("Time")
ax.set_ylabel("Amplitude")
ax.legend()

# Animation function
def animate(frame):
    analog_line.set_data(t[:frame], analog_signal[:frame])
    digital_samples.set_data(sample_points_x[:frame], sample_points_y[:frame])
    return analog_line, digital_samples

# Create the animation
ani = FuncAnimation(fig, animate, frames=len(t), blit=True, interval=50)

#Save the animation (optional)
ani.save('analog_to_digital_animation.mp4', writer='ffmpeg')

plt.show()
```
