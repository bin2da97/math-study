## weighted sum(가중합 x와 w의 선형결합/linear combination/내적)에 bias를 더한 결과를 활성화함수(activation function의 입력으로 들어감) 
- 활성화함수
       -sigmoid, 
       -tanh
       -계단함수(step function)
       -Relu가 있고 최근에는 Relu가 가장 많이 사용됨
우리가 찾고자 하는것은 x에 대응되는 weight(각 x에 대응되는 계수)와 bias를 찾는것임

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def step_function(x): #0보다 큰 구간에서 값을 반환하고 0보다 작은 구간에서는 0
    return np.array(x>0, dtype=np.int8)
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x=np.arange(-10,10,0.01)
y=sigmoid(x)
plt.plot(x,y)
```

```python
x=np.arange(-10,10,0.01)
y=relu(x)
plt.ylim(-0.1, 5.5)
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(-10,10,0.01)
y=step_function(x)
plt.ylim(-0.1,1.1)
plt.plot(x,y)
plt.show()
```
