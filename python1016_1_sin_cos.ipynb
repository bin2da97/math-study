```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def dri(f, x,h=0.00001):
    return (f(x+h)-f(x))/h
```

```python
x=np.arange(0, 6, 0.1)#0에서 6까지 0.1간격으로 생성
y1=np.sin(x)
y2=np.cos(x)
#그래프그리기
plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle="--" ,label='cos')#cos함수는 점선으로 그리기
plt.xlabel("x")#x축이름
plt.ylabel("y")#y축이름
plt.title('sin & cos')
plt.legend(loc='upper right')
plt.show
```

```python
x=np.arange(-10, 10, 0.01)#0에서 6까지 0.1간격으로 생성
y=3*x**2
y1=3*(x-3)**2+5
#그래프그리기
plt.plot(x, y, label='y')
plt.plot(x,y1, label='y1')
plt.ylim(-1,100)
plt.title('sin & cos')
plt.legend(loc='upper right')
plt.show
```

```python
x=np.arange(0, 6, 0.1)#0에서 6까지 0.1간격으로 생성
y1=np.sin(x)
y2=np.cos(x)
y1_dri=dri(np.sin, x)+0.1#미분하면 y2와 중첩되기때문에 y축으로 0.1만큼이동
y2_dri=dri(np.cos, x)
#그래프그리기
plt.plot(x, y1, label='sin')
plt.plot(x, y2, label='cos')
plt.plot(x, y2, linestyle="--" ,label='cos')#cos함수는 점선으로 그리기
plt.plot(x,y1_dri, label='y1_dri', linestyle="--" )#sin함수미분은 cos
plt.plot(x, y2_dri, label='y2_dri', linestyle="--")#cos함수미분은 -sin
plt.xlabel("x")#x축이름
plt.ylabel("y")#y축이름
plt.title('sin & cos$ sin_der &cos_der' )
plt.legend(loc='upper right')
plt.show
```

## 평행이동
$$ y=3x^{3} \rightarrow y= 3(x+7)^{3}+2 $$
$$ y=3x^3을 어떻게 이동하는가? $$
$$ y=3x^{3} \rightarrow y= -(3(x+7)^{3}+2) $$
$$ y= -(3(x+7)^{3}+2)의 그래프는 어떻게 되는가? $$ 

```python
x=np.arange(-10, 3,0.01)#0에서 6까지 0.1간격으로 생성
y_ori=3*x**3
y_moving=3*(x+7)**3+2
y_reflection=-(3*(x+7)**3+2)
#그래프그리기
plt.plot(x, y_ori, label='y_ori')
plt.plot(x,y_moving, label='y_moving')
plt.plot(x,y_reflection, label='y_reflection')
plt.ylim(-10,10)
plt.show
```
