```python
import sys
from PyQt6.QtWidgets import (QApplication,QWidget,QLabel,QPushButton)
from PyQt6.QtCore import Qt
import matplotlib.pyplot as plt
import numpy as np
import math

class MainWindow(QWidget):
    def __init__(self):
        super().__init__()
        self.initializeUI()
        self.list_func1=[lambda x: x ** 2 + 3 * x + 5, lambda x: 2 * x + 5 * x + 2, lambda x: 3 * x ** 3 + x + 1]


    def initializeUI(self):
        """
        어플리케이션 UI 설정
        """
        self.setGeometry(100,100,400,300)
        self.setWindowTitle("QPushButton 예제")
        self.setUpMainWindow()
        self.show()

    def setUpMainWindow(self):
        """
        메인 윈도우에 위젯을 생성후 배치
        """
        self.times_pressed=0
        self.times_pressed1=0
        self.times_pressed2=0
        self.times_pressed3=0
        self.times_pressed4=0

        self.name_label=QLabel("버튼을 누르지 말아주세요 아픔", self)
        self.name_label.setAlignment(Qt.AlignmentFlag.AlignCenter)
        self.name_label.move(60,30)

        self.button = QPushButton("나를 눌러주세요",self)
        self.button.move(60,70)
        self.button.clicked.connect(self.buttonClicked)

        self.button1 = QPushButton("구구단", self)
        self.button1.move(60, 100)
        self.button1.clicked.connect(self.buttonClicked1)

        self.button2 = QPushButton("함수 미분", self)
        self.button2.move(60, 130)
        self.button2.clicked.connect(self.buttonClicked2)

        self.button2 = QPushButton("함수 미분 선생님", self)
        self.button2.move(170, 130)
        self.button2.clicked.connect(self.buttonDiifClicked)

        self.button3 = QPushButton("테일러", self)
        self.button3.move(170, 70)
        self.button3.clicked.connect(self.buttonClicked3)

    def buttonClicked(self):
        """
            버튼이 눌리면 호춣되는 HANDLER
            위젯이 텍스트가 변경되면 위젯의 크기와 위치를 변경하고 이벤트에
            대한 윈도우 종료
        """
        self.times_pressed+=1
        if self.times_pressed==1:
            self.name_label.setText("나를 왜 누르는가요?")
        if self.times_pressed==2:
            self.name_label.setText("경고중")
            self.button.setText("행운을")
            self.button.adjustSize()
            self.button.move(70,70)
        if self.times_pressed==3:
            print("윈도우 닫")
            self.close()

    def buttonClicked1(self):
        self.times_pressed+=1

        if self.times_pressed==1:
            self.name_label.setText("구구단")
            for i in range(9):
                print(i+1,'단')
                for j in range(9):
                    print(i+1,'x',(j+1),'=',(i+1)*(j+1))
        # if self.times_pressed2==10:
        #     self.times_pressed2=0

    def derivative(self, f , x, h=0.00001):
        #print(f)
        return (f(x+h)-f(x))/h
    def derivative1(self, f , x, h=0.00001):
        return np.array((f(x+h)-f(x))/h)

    def buttonClicked2(self):
        list_func=[lambda x: x ** 2 + 3 * x + 5, lambda x: 2 * x + 5 * x + 2, lambda x: 3 * x ** 3 + x + 1]
        x = np.arange(-10, 10, 0.01)
        y_derivative = self.derivative(list_func[self.times_pressed2-1], x)
        print(y_derivative)

        plt.plot(x, self.list_func1[self.times_pressed2-1](x))
        plt.plot(x, y_derivative)
        plt.draw()
        plt.pause(0.0001)
        plt.clf()
        self.times_pressed2+=1

        if self.times_pressed2==3:
            self.times_pressed2=0


    def buttonDiifClicked(self):
        print('미분')
        x=np.arange(-10,10,0.01)
        for i in range(len(self.list_func1)):
            if i==self.times_pressed4:
                y_deri = self.derivative(self.list_func1[i],x)
                plt.plot(x,self.list_func1[i](x))
                plt.plot(x,y_deri)
                plt.draw()
                plt.pause(0.0001)
                plt.clf()
        self.times_pressed4+=1
        if self.times_pressed4==len(self.list_func1):
             self.times_pressed4=0


    def func_cos(self,x,n):
        cos_approx = 0
        for i in range(n):
            coef = (-1) ** i
            num = x ** (2 * i)
            denom = math.factorial(2 * i)
            cos_approx += (coef) * ((num) / (denom))
        return cos_approx
    
    def buttonClicked3(self):
        angles = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
        t_cos = np.cos(angles)
        fig, ax = plt.subplots()
        ax.plot(angles, t_cos)
        for i in range(self.times_pressed3):
            p_cos = [self.func_cos(angle, i) for angle in angles]
            ax.plot(angles, p_cos)
        ax.set_ylim([-7, 4])
        legend_list = ['cos() function']
        for i in range(self.times_pressed3):
            legend_list.append(f'Taylor Series -{i} terms')
            ax.legend(legend_list, loc=3)
        plt.show()
        self.times_pressed3+=1



if __name__=='__main__':
    app= QApplication(sys.argv)
    window=MainWindow()
    sys.exit(app.exec())
```

```python
import sys
from PyQt6.QtWidgets import (QApplication, QWidget, QLabel, QPushButton, QTextEdit, QMessageBox, QFileDialog)
from PyQt6.QtCore import Qt
import numpy as np
import matplotlib.pyplot as plt


class NotePad(QWidget):

    def derivative(self, f, x, h=1e-5):
        return np.array((f(x + h) - f(x)) / h)

    def __init__(self):
        super().__init__()
        self.list_func = [lambda x: x ** 2, lambda x: x ** 3, lambda x: x ** 4]
        self.data = ['1번', '2번', '3번']
        self.data_der = ['1번 미분', '2번 미분', '3번 미분']
        self.data_total = [self.data, self.data_der]
        self.initializeUI()

    def initializeUI(self):
        """
            어플리케이션 UI 설정
         """

        self.setGeometry(100, 100, 300, 400)
        self.setWindowTitle("Notepad GUI")
        self.notepadWidgets()
        self.show()

    def notepadWidgets(self):
        """
            편집 메뉴를 위한 푸시 버튼 생성
        """
        save_button = QPushButton("저장", self)
        save_button.move(80, 20)
        save_button.clicked.connect(self.saveText)  ## 문1) 2개의 event handler 작성
        # 텍스트 편집 field
        self.text_field = QTextEdit(self)
        self.text_field.resize(280, 330)
        self.text_field.move(10, 60)

    def saveText(self):
        """
            "저장" 버튼이 눌리면 텍스트 편집 field에 있는 내용을 저장할 지  표시 대화 창
        """
        print("저장버튼이 눌렸어요")
        notepad_text = self.text_field.toPlainText()
        splited = notepad_text.split("\n")
        # v = [i if i in self.data_total[1] else i for i in splited]
        # print('미분 포함:', v)

        contain_string_dir = []
        not_contain_string_dir=[]
        for i in splited:
            if i in self.data_total[1]:
                contain_string_dir.append(i)
            else:
                not_contain_string_dir.append(i)

        # contain_string_dir = [i for i in splited if i in self.data_total[1]]
        # not_contain_string_dir = [i for i in splited if i not in self.data_total[1]]
        list_zip = zip(self.list_func, self.data)
        list_der_zip = zip(self.list_func, self.data_der)
        x = np.arange(-10, 10, 0.01)
        for f, d in list_zip:
            if d in not_contain_string_dir:
                plt.plot(x, f(x))

        for f, d in list_der_zip:
            if d in contain_string_dir:
                plt.plot(x, self.derivative(f, x))
        plt.show()


if __name__ == '__main__':
    app = QApplication(sys.argv)
    window = NotePad()
    sys.exit(app.exec())
```
