```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
np.arccos(2/5)
```

```python
actual_values = [2.5, 3.0, 4.0, 4.5, 5.0]
predicted_values = [2.7, 2.9, 3.8, 4.2, 5.1]
```

```python
def mean_squared_error(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

mse = mean_squared_error(actual_values, predicted_values)
print("Mean Squared Error:", mse)
```

```python
# 실제 값과 예측 값
actual_values = [2.5, 3.0, 4.0, 4.5, 5.0]
predicted_values = [2.7, 2.9, 3.8, 4.2, 5.1]

# 차이의 제곱 계산
squared_errors = (np.array(actual_values) - np.array(predicted_values)) ** 2

# MSE 계산
mse = np.mean(squared_errors)

# 시각화
plt.figure(figsize=(10, 6))

# 실제 값과 예측 값 시각화
plt.scatter(actual_values, predicted_values, color='blue', label='Data points')
plt.plot(actual_values, actual_values, color='red', label='Perfect prediction')

# 차이의 제곱 시각화
for i, (actual, predicted) in enumerate(zip(actual_values, predicted_values)):
    plt.plot([actual, actual], [actual, predicted], color='gray')
    plt.plot(actual, predicted, 'go', label=f'Data point {i+1}')

plt.title('MSE Calculation Visualization')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.grid()
plt.show()

print("Mean Squared Error:", mse)
```
