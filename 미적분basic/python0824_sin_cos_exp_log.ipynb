$$ \lim_{h\rightarrow 0} \frac{f(x+h)-f(x)}{h} $$

```python
import os
```

```python
def derivative(f,x):
    h=0.00001
    return [(f(i+h)-f(i))/h for i in x]
```

```python
fss=['cos','sin','exp','log','2x^3+3','x^2']
for i,f in enumerate(zip(fn,fss)):
    plt.subplot(6,2,i*2+1)
    plt.title(f'{f[1]} function')
    plt.ylabel(f'{f[1]} value')    
    plt.plot(x,result[i])    
    plt.subplot(6,2,i*2+2)
    plt.title(f'{f[1]} derivative function')
    plt.ylabel(f'{f[1]} value') 
    plt.plot(x,derivative(f[0],x))    
```
