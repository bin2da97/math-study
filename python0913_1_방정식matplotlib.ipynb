```python
#y = x^2+5 포물선의 방정식 그리기
x=np.arange(-10,10,0.01)
y=[i**2+5 for i in x] #y=x를 y축(x=0) 대칭한 직선의 방정식
plt.plot(x,y)
```

```python
#문제2 y=x^2과 x^2+5의 관계 
#  y축으로 5만큼 이동

#문제3 y=(x+5)^2+11=> 그래프 그리기
x=np.arange(-10,10,0.01)
y=[(i+5)**2+11 for i in x] #y=x를 y축(x=0) 대칭한 직선의 방정식
plt.plot(x,y)
```
