```python
import numpy as np
import cv2


def camera_in():
    cap = cv2.VideoCapture(0)

    if not cap.isOpened():
        print("Camera open failed!")
        return

    print('Frame width:', int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)))
    print('Frame height:', int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        inversed = ~frame

        cv2.imshow('frame', frame)
        cv2.imshow('inversed', inversed)

        if cv2.waitKey(10) == 27:
            break

    cv2.destroyAllWindows()


def video_in():
    cap = cv2.VideoCapture('images/stopwatch.avi')

    if not cap.isOpened():
        print("Video open failed!")
        return

    print('Frame width:', int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)))
    print('Frame height:', int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
    print('Frame count:', int(cap.get(cv2.CAP_PROP_FRAME_COUNT)))

    fps = cap.get(cv2.CAP_PROP_FPS)
    print('FPS:', fps)

    delay = round(1000 / fps)

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        inversed = ~frame

        cv2.imshow('frame', frame)
        cv2.imshow('inversed', inversed)

        if cv2.waitKey(delay) == 27:
            break

    cv2.destroyAllWindows()


def camera_in_video_out():
    cap = cv2.VideoCapture(0)

    if not cap.isOpened():
        print("Camera open failed!")
        return

    w = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    h = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    fourcc = cv2.VideoWriter_fourcc(*'DIVX') # *'DIVX' == 'D', 'I', 'V', 'X'
    delay = round(1000 / fps)

    outputVideo = cv2.VideoWriter('output.avi', fourcc, fps, (w, h))

    if not outputVideo.isOpened():
        print('File open failed!')
        return

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        inversed = ~frame

        outputVideo.write(inversed)

        cv2.imshow('frame', frame)
        cv2.imshow('inversed', inversed)

        if cv2.waitKey(delay) == 27:
            break

    cv2.destroyAllWindows()


if __name__ == '__main__':
    camera_in()
    video_in()
    camera_in_video_out()
```

```python
from __future__ import print_function
import numpy as np
import imutils
import time
import cv2




output = 'images/example.avi'
video = 'images/stopwatch.avi' #읽어드릴 파일
fps = 33
codec ='MJPG'


vs = cv2.VideoCapture(video)
time.sleep(2.0)
#  FourCC, video writer, frame 차원, zeros배열로 초기화
fourcc = cv2.VideoWriter_fourcc(*codec) #저장하기 위한 객체 생성
writer = None
(h, w) = (None, None) #세로(height),가로(width)
zeros = None
cnt=0
    
while True:
    cnt+=1
    if cnt%33==0:
            ret, frame = vs.read() #입력 비디오 파일로부터 한 프레임 이미지씩 읽어드린다
            print(cnt, frame.shape)
    if ret==True:
        frame = imutils.resize(frame, width=300) #크기를 재조정
    # check if the writer is None
    else:
        break
    if writer is None:

        (h, w) = frame.shape[:2]
        writer = cv2.VideoWriter(output, fourcc, fps,
            (w * 2, h * 2), True)
        zeros = np.zeros((h, w), dtype="uint8")

    (B, G, R) = cv2.split(frame)
    A=cv2.merge([B*0.5,G*0.3,R*0.2])
    R = cv2.merge([zeros, zeros, R])
    G = cv2.merge([zeros, G, zeros])
    B = cv2.merge([B, zeros, zeros])

#     print(B[:20],G[:20],R[:20])
    output = np.zeros((h * 2, w * 2, 3), dtype="uint8")
    output[0:h, 0:w] = A
    output[0:h, w:w * 2] = R
    output[h:h * 2, w:w * 2] = G
    output[h:h * 2, 0:w] = B

    writer.write(output)

    cv2.imshow("Frame", frame)
    cv2.imshow("Output", output)
    key = cv2.waitKey(1) & 0xFF

    if key == ord("q") or key == ord("Q") :
        break

print("[INFO] cleaning up...")
cv2.destroyAllWindows()
vs.release()
writer.release()
```

```python
TIMER=int(20)
cap = cv2.VideoCapture('images/stopwatch.avi')
while True:
    ret,img = cap.read()
    cv2.imshow('a',img)
    k=cv2.waitKey(125)
    if k==ord('q'):
        prev=time.time()
        while TIMER>=0:
            ret,img = cap.read()
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(img,str(TIMER),(200,250),FONT,7(0,255,255),4,cv2.LINE_AA)
            cv2.imshow('a',img)
            cv2.waitKey(125)
            cur = time.time()
            if cur-prev>=1:
                prev=cur
                TIMER=TIMER-1
            else:
                ret,img = cap.read()
                cv2.imshow('a',img)
                cv2.waitKey(2000)
                cv2.imwrite('images/camera.jpg',img)
    elif k==27:break
cap.release()
cv2.destroyAllWindows()

        
```
