```python
def deg_to_rad(i): #각도 => 라디안
    return i*np.pi*2/360
```

```python
def rad_to_deg(i): #라디안 => 각도 
    return (360*i)/(np.pi*2)
```

$$ \log_{a}{b}=\frac{\log_{c}{b}}{\log_{c}{a}} $$
$$ \log_{10}{\frac{1}{10}}=-1 $$
$$ \log_{\frac{1}{10}}{10}=-1 $$
$$ \log_{e}{\frac{1}{e}}=-1 $$
$$ \log_{\frac{1}{e}}{e}=-1 $$

```python
cos_a=[3/10,5/10,2/5,3/7,1/2]
sin_a=[np.sqrt(91)/10,np.sqrt(75)/10,np.sqrt(21)/5,np.sqrt(40)/7,np.sqrt(3)/2]

sum_result=[i**2+j**2 for i,j in zip(cos_a,sin_a)]
[0 if abs(i-1)<0.00001 else 1 for i in sum_result]
```

$$ cos^{2}\theta+sin^{2}\theta=1 $$
제곱해서 2개를 더해서 1이 되려면 하나는 $\frac{1}{\sqrt{2}}$이
되어야 하므로 제곱해서 $\frac{1}{2}$이 되기 위해서
원래 값은 $\frac{1}{\sqrt{2}}=\frac{\sqrt{2}}{2}$
가 되고 서로 같으므로 cos,sin 값이 같으려면 이등변 삼각형이므로 각도는 45도

```python
a=[1,2,3,7,5]
b=[2,4,8,2,-1]
c=[-3,-6,-2,-7,5]
# a^2+b^2+c^2 <100 조건 만족하는 데이터 찾기
for x,y,z in zip(a,b,c):
    if x**2+y**2+z**2<100:
        print(x,y,z)
```

```python
result=[(x,y,z)  for x,y,z in zip(a,b,c) if x**2+y**2+z**2<100]
result
```
