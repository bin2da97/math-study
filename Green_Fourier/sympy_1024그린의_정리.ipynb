## 선 적분과 2중 적분의 결과가 같다.

$$ \oint_{\partial D} P dx + Q dy = \int\int_{D}(\frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} ) dx dy \rightarrow \oint = 폐곡선을 따라 적분,$$ $$\partial D= 경계선을 따라 적분(적분 구간), \int_{D} 에서 D는 면적 적분 $$

$$ \oint_{\partial D} P dx + Q dy  = \int\int_{D} 
    (\frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y}) 
dx dy$$

$$ F(x,y) = (x-y)i + xj \rightarrow i = x 축방향으로 크기가 1인, j=y축 방향으로 크기가 1인 단위벡터$$

적분 영역 (D)는 원점을 중심으로 반지름이 1인 단위원
$$ r(t) = (cost)i+(sint)j $$ ,곡선은 r로 표시

## 우변 
벡터 함수의 편미분들을 계산
$$\frac{\partial F_{1}}{\partial y} = -1  $$

$$\frac{\partial F_{2}}{\partial x} = 1  $$ 

$$ \int\int_{D}(\frac{\partial F_{2}}{\partial x} -  \frac{\partial F_{1}}{\partial y} ) dA $$
= $$ \int\int_{D}(1-(-1)) dx dy =2 \int\int_{D} dx dy = 2\pi $$

```python
import sympy as sp

# 변수 및 반지름 정의
x, y, r = sp.symbols('x y r')
radius = sp.Symbol('r', positive=True)  # 반지름은 양수로 가정합니다.

# 원의 방정식
circle_eq = x**2 + y**2

# 이중적분 표현
area_integral = sp.integrate(sp.integrate(1, (x, -sp.sqrt(r**2 - y**2), sp.sqrt(r**2 - y**2))), (y, -r, r))

# 결과 계산
result = area_integral.simplify()

print("원의 면적:", result)
```

```python
import numpy as np
```

```python
# 변수 정의
x, y = sp.symbols('x y')

# 피적분 함수 정의
f = x**2 + y**2

# 적분 범위 정의 (예: 사각형 영역 [0, 1] x [0, 2])
x_min, x_max = 0, 1
y_min, y_max = 0, 2

# 2중 적분 계산
result = sp.integrate(sp.integrate(f, (x, x_min, x_max)), (y, y_min, y_max))

print("2중 적분 결과:", result)
abs(result-np.pi)<0.7
```

## 좌변
원위의 점을 따라가며 형성되는 벡터 함수 값을 찾는다.

<b>선적분</b>에 필요한 <b>연쇄법칙(chain rule)</b>을 계산한다.

$$ dx = d(cost) = -sint dt,dy=d(sint) =cost dt $$

반시계 방향으로 회전하며 우변을 적분한다.

$$ \oint_{C} F_{1} dx + F_{2} dy = \int_{t=0}^{t=2\pi}(cost - sint)(-sint dt)+(cost)(costdt) 
= \int_{t=0}^{t=2\pi}(-sintcost + 1 ) dt = 2\pi  $$

```python
t_values = np.linspace(0, 2*np.pi, 1000)  # 0부터 2π까지 t 값을 생성합니다.
# 좌변과 우변을 계산합니다.
left_side = np.sin(t_values) * np.cos(t_values)
right_side = 1/2 * np.sin(2 * t_values)

# 두 변이 동일한지 확인합니다.
are_equal = np.allclose(left_side, right_side)

if are_equal:
    print("두 변은 동일합니다.")
else:
    print("두 변은 동일하지 않습니다.")
```
