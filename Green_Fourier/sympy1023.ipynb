## 오일러 공식
$$ e^{j\theta}= cos(\theta) + jsin(\theta) $$
반지름이 1인 단위원을 생각하고 실수축은 cos, 허수측은 sin 생각

2개의 벡터를 곱하는 것의 크기는 각각의 벡터의 크기는 곱하고 각도는 더한다.

$$ \overrightarrow{a} \cdot \overrightarrow{b} = 
    |\overrightarrow{a}| \cdot |\overrightarrow{b}| e^{j(\theta_{1}+\theta_{2})}
$$
여기서 $\theta_{1}$은 x축을 기준으로 $\overrightarrow{a}$의 벡터까지 <br>
반시계 방향의 각도이고 $\theta_{2}$은 x축을 기준으로 $\overrightarrow{b}$의 벡터까지의 각도이다

```python
import numpy as np
```

```python
a = np.array([1,2])
b = np.array([2,1])
c = np.array([4,3])
```

```python
import math

def calculate_angle(vector):
    # 벡터의 크기 계산
    magnitude = math.sqrt(vector[0]**2 + vector[1]**2)
    
    # x축과 이루는 각도 계산 (라디안 단위)
    if vector[0] == 0:
        if vector[1] > 0:
            angle = math.pi / 2
        elif vector[1] < 0:
            angle = -math.pi / 2
        else:
            angle = 0  # 벡터가 원점에 있을 때
    else:
        angle = math.atan(vector[1] / vector[0])
    
    # 라디안을 도 단위로 변환
    angle_degrees = math.degrees(angle)
    
    return magnitude, angle_degrees

# 테스트
vector = (3, 4)  # (x, y) 좌표가 주어짐
magnitude, angle = calculate_angle(vector)
print(f"크기: {magnitude}, 각도: {angle}도")
```

```python
a_m,a_angle=calculate_angle(a)
b_m,b_angle=calculate_angle(b)
c_m,c_angle=calculate_angle(c)
```

```python
print('a 벡터의 크기 :',a_m,'각도 : ', a_angle)
print('b 벡터의 크기 :',b_m,'각도 : ', b_angle)
print('c 벡터의 크기 :',c_m,'각도 : ', c_angle)

a_m*b_m*c_m
```
