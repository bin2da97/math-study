# 푸리에 변환
- 시간의 신호를 주파수의 신호로 변환

```python
y_sin = np.sin(x)
plt.plot(x,y_sin)
```

$$ 0.6\sin(2\pi f_{1} \cdot t) + 0.6\cos(2\pi f_{2} \cdot t+\frac{\pi}{2}) $$

$ \omega = 2\pi f \rightarrow \omega $ 는 각속도 ,f = 주파수(frequency)

$$ 0.6\sin(\omega_{1} \cdot t) + 0.6\cos(\omega_{2} \cdot t+\frac{\pi}{2}) $$
$$ \cos(\theta)=\cos(\omega t) = \cos(2\pi f t)  = $$

$$ 속도(v) = \frac{s}{t} \rightarrow 각속도(\omega) = \frac{\theta}{t} $$
$$ \theta = \omega t $$

```python
fs = 100
t = np.arange(0,3,1/fs)
f1=35 
f2=10 
# 주파수 35이고 진폭(높이)이 0.6인 사인파와 주파수 10인 cos파를
# 0도 이동한 신혹의 진폭(높이)이 3인 신호의 합성
signal = 0.6*np.sin(2*np.pi*f1*t)+3*np.cos(2*np.pi*f2*t+np.pi/2)
fft=np.fft.fft(signal) / len(signal)
fft_magnitude = abs(fft)
# fft_magnitude
plt.subplot(2,1,1)
plt.plot(t,signal)
plt.grid()

plt.subplot(2,1,2)
plt.stem(fft_magnitude)
plt.ylim(0,2.5)
plt.grid()
plt.show()
```

```python
fs = 100
t = np.arange(0,3,1/fs)
f1 =35 
f2 =10 
f3 = 100
# 주파수 35이고 진폭(높이)이 0.6인 사인파와 주파수 10인 cos파를
# 0도 이동한 신혹의 진폭(높이)이 3인 신호의 합성 
signal = 0.6*np.sin(2*np.pi*f1*t)+3*np.cos(2*np.pi*f2*t+np.pi/2)+1*np.cos(2*np.pi*f3*t-np.pi/2) 
fft=np.fft.fft(signal) / len(signal)
fft_magnitude = abs(fft)
# fft_magnitude
plt.subplot(2,1,1)
plt.plot(t,signal)
plt.grid()

plt.subplot(2,1,2)
plt.stem(fft_magnitude)
plt.xlim(-500,500)
plt.ylim(0,2.5)
plt.grid()
plt.show()
```
