$$ x(t) = t^2-3 $$
$$ y(t) = 2t+1 $$
$$ -2 \lt t \lt 3 $$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# t의 범위를 정의합니다.
t = sp.symbols('t')
t_values = np.linspace(-2, 3, 400)  # NumPy의 linspace를 사용하여 등간격으로 t 값을 생성합니다.

# x(t)와 y(t)를 정의합니다.
x = t**2 - 3
y = 2*t + 1

# t_values를 사용하여 x와 y 값을 계산합니다.
x_values = [x.subs(t, val) for val in t_values]
y_values = [y.subs(t, val) for val in t_values]

# 그래프를 그립니다.
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Parametric Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parametric Curve: x(t) = t^2 - 3, y(t) = 2t + 1')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# t의 범위를 정의합니다.
t = sp.symbols('t')
t_values = np.linspace(0, 2*np.pi, 400)  # NumPy의 linspace를 사용하여 등간격으로 t 값을 생성합니다.

# x(t)와 y(t)를 정의합니다.
x = 4*sp.cos(t)
y = 3*sp.sin(t)

# t_values를 사용하여 x와 y 값을 계산합니다.
x_values = [x.subs(t, val) for val in t_values]
y_values = [y.subs(t, val) for val in t_values]

# 그래프를 그립니다.
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Parametric Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parametric Curve: x(t) = t^2 - 3, y(t) = 2t + 1')
plt.grid(True)
plt.legend()
plt.show()
```
