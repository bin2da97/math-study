디랙 델타 함수(Dirac Delta Function)
- 신호처리분야에서 임펄스(impulse) 함수

$$ \begin{equation}
  \delta(x)=\left\{
  \begin{array}{@{}ll@{}}
    \infty, & \text{if}\ x=0 \\
    0  & \text{if}\  x \neq 0  
  \end{array}\right.
\end{equation} 
$$

$$  
    \int_{-\infty}^{\infty} \delta(x) dx = 1
$$

$$  
  \delta(x-\alpha) = \frac{1}{2\pi} \int_{-\infty}^{\infty} e^{ip(x-\alpha)} dp
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# 샘플링 주파수 (Hz)
sampling_frequency = 50  # 샘플링 주파수를 높게 설정하면 더 정밀한 샘플링이 가능합니다.

# 시간 벡터 생성 (1초 동안)
t = np.arange(0, 1, 1/sampling_frequency)

# sin(2*pi*t) 함수 샘플링
sampled_signal = np.sin(2 * np.pi * t)

# 결과 플롯
plt.stem(t, sampled_signal, 'r', basefmt=" ")
plt.plot(t, np.sin(2*np.pi*t), 'b')
plt.title('Sampling of sin(2*pi*t) using Delta Function')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 샘플링 주파수 (Hz)
sampling_frequency = 100  # 샘플링 주파수를 높게 설정하면 더 정밀한 샘플링이 가능합니다.

# 시간 벡터 생성 (1초 동안)
t = np.arange(0, 1, 1/sampling_frequency)

# sin(2*pi*t) 함수 샘플링
sampled_signal = 8*np.sin(2 * np.pi * t)*4*np.sin(4*np.pi*t)

# 결과 플롯
plt.stem(t, sampled_signal, 'r', basefmt=" ")
plt.plot(t, 8*np.sin(2 * np.pi * t)*4*np.sin(4*np.pi*t), 'b')
plt.title('Sampling of sin(2*pi*t) using Delta Function')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 시간 범위 설정 (예: 1초 동안, 샘플링 주기 0.001초)
sampling_frequency = 1000
t = np.arange(0, 1, 1/sampling_frequency)

# 원래의 아날로그 신호 (예: sin(2πt) 함수)
analog_signal = np.sin(2 * np.pi * t)

# ADC: 아날로그 신호를 디지털로 샘플링
digital_signal = analog_signal

# DAC: 디지털 신호를 아날로그로 복원
reconstructed_analog_signal = digital_signal

# 결과 플롯
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, analog_signal)
plt.title('원래 아날로그 신호')
plt.xlabel('시간 [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(t, digital_signal, 'r', basefmt=" ")
plt.title('ADC 결과 (디지털 신호)')
plt.xlabel('시간 [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, reconstructed_analog_signal)
plt.title('DAC 결과 (아날로그 신호 재생성)')
plt.xlabel('시간 [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 시간 범위 설정 (예: 1초 동안, 샘플링 주기 0.001초)
sampling_frequency = 1000
t = np.arange(0, 1, 1/sampling_frequency)

# 원래의 아날로그 신호 (예: sin(2πt) 함수)
analog_signal = np.sin(2 * np.pi * t)

# ADC 파라미터 설정
quantization_bits = 4  # ADC의 비트 수 (양자화 레벨)
quantization_levels = 2**quantization_bits  # 양자화 레벨 수

# 아날로그 신호를 ADC를 통해 디지털로 양자화
quantized_signal = np.round((analog_signal + 1) * (quantization_levels - 1) / 2)  # 0부터 양자화 레벨까지의 정수 값

# 양자화된 신호를 다시 아날로그로 DAC
reconstructed_analog_signal = quantized_signal * 2 / (quantization_levels - 1) - 1  # -1부터 1까지의 아날로그 신호

# 결과 플롯
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, analog_signal)
plt.title('original analog signal')
plt.xlabel('time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(t, quantized_signal, 'r', basefmt=" ")
plt.title(f'ADC result (quantization level: {quantization_levels})')
plt.xlabel('time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, reconstructed_analog_signal)
plt.title('DAC result (analog signal reconstruction)')
plt.xlabel('time [s]')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
