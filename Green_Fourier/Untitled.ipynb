```python
import sympy as sp
```

```python
x,y=sp.symbols('x y')
eq1=sp.Eq(2*x+y,3)
eq2=sp.Eq(x-y,7)
```

```python
pp1 = sp.plot_implicit(eq1)# implicit 음함수 y=2x 가 아니라, 2x+y+3=0 이런거
pp2 = sp.plot_implicit(eq2)
pp1.extend(pp2)
pp1.save('figs/ex4a.pdf')
```

### row echelon form(REF) 으로 문제 풀기
$$ x+2y+3z=12 $$
$$ 2x-y+5z=3 $$
$$ 3x+3y+6z=21 $$

$$ \begin{bmatrix}
1 & 2 & 3 & 12 \\
 2& -1 & 5 &3  \\
3 & 3 & 6 & 21 
\end{bmatrix} $$
