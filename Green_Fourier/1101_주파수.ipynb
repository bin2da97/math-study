```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

#  10    
num_points = 360 // 10

#   
angles = np.radians(np.arange(0, 360, 10))

#    
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_aspect('equal', adjustable='box')

#      
all_images = []

for v in range(1, 6):  #  1 5 
    #  x, y  
    x = v * np.cos(angles)
    y = v * np.sin(angles)

    #       
    images = []
    for i in range(num_points):
        ax.clear()
        
        # Draw all circles with different radii
        for r in range(1, v + 1):
            circle_x = r * np.cos(angles)
            circle_y = r * np.sin(angles)
            ax.plot(circle_x, circle_y, 'b')

        #    
        ax.plot([x[i]], [y[i]], 'ro')

        #     
        ax.set_title(f"Angle: {i * 10} degrees, Radius: {v}")

        #     
        plt.tight_layout()
        fig.canvas.draw()
        img = np.array(fig.canvas.renderer.buffer_rgba())
        images.append(Image.fromarray(img))

    #         
    all_images.extend(images)

#    GIF  
all_images[0].save(f'unit_circle_animation_with_multiple_circles.gif', save_all=True, append_images=all_images[1:], duration=100)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

#  10    
num_points = 360 // 10

#   
angles = np.radians(np.arange(0, 360, 10))

#    
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

#     .
ax1.set_xlim(-1.2, 1.2)
ax1.set_ylim(-1.2, 1.2)
ax1.set_aspect('equal', adjustable='box')
ax1.set_title('Rotating Circle')

#    cos sin   .
ax2.set_xlim(0, 360)
ax2.set_ylim(-1.5, 1.5)
ax2.set_title('cos and sin Waveforms')

#      
all_images = []

for i in range(num_points):
    ax1.clear()
    ax2.clear()

    #       
    rotated_circle_x = np.cos(angles - np.radians(i * 10))
    rotated_circle_y = np.sin(angles - np.radians(i * 10))
    ax1.plot(rotated_circle_x, rotated_circle_y, 'b')

    # cos    ( )
    cos_wave_x = np.arange(360)
    cos_wave_y = np.cos(np.radians(i * 10 - cos_wave_x))
    ax2.plot(cos_wave_x, cos_wave_y, 'r', label='cos Waveform')

    # sin    ( )
    sin_wave_y = np.sin(np.radians(i * 10 - cos_wave_x))
    ax2.plot(cos_wave_x, sin_wave_y, 'g', label='sin Waveform')

    #   (  )
    if i == 0:
        ax2.legend()

    #     
    plt.tight_layout()
    fig.canvas.draw()
    img = np.array(fig.canvas.renderer.buffer_rgba())
    all_images.append(Image.fromarray(img))

#    GIF  
all_images[0].save('rotating_circle_with_fixed_scale_cos_sin.gif', save_all=True, append_images=all_images[1:], duration=100)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

#  10    
num_points = 360 // 10

#   
angles = np.radians(np.arange(0, 360, 10))

#    
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

#     .
ax1.set_xlim(-1.2, 1.2)
ax1.set_ylim(-1.2, 1.2)
ax1.set_aspect('equal', adjustable='box')
ax1.set_title('Rotating Circle')

#    cos sin   .
ax2.set_xlim(0, 360)
ax2.set_ylim(-1.2, 1.2)  #      .
ax2.set_title('cos and sin Waveforms')

#       
all_images = []

for frequency in range(1, 25, 5):
    for i in range(num_points):
        ax1.clear()
        ax2.clear()

        #       
        rotated_circle_x = 1.0 * np.cos(frequency * angles - np.radians(i * 10))
        rotated_circle_y = 1.0 * np.sin(frequency * angles - np.radians(i * 10))
        ax1.plot(rotated_circle_x, rotated_circle_y, 'b')

        # cos    ( )
        cos_wave_x = np.arange(360)
        cos_wave_y = np.cos(np.radians(i * 10 - frequency * cos_wave_x))
        ax2.plot(cos_wave_x, cos_wave_y, 'r', label='cos Waveform')

        # sin    ( )
        sin_wave_y = np.sin(np.radians(i * 10 - frequency * cos_wave_x))
        ax2.plot(cos_wave_x, sin_wave_y, 'g', label='sin Waveform')

        #   (  )
        if i == 0:
            ax2.legend()

        #     
        plt.tight_layout()
        fig.canvas.draw()
        img = np.array(fig.canvas.renderer.buffer_rgba())
        all_images.append(Image.fromarray(img))

#    GIF  
all_images[0].save('rotating_circle_and_cos_sin_with_fixed_scale.gif', save_all=True, append_images=all_images[1:], duration=100)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft

#    
t = np.linspace(0, 1, 1000, endpoint=False)  # 1  1000 
signal = np.sin(10 * np.pi * 10 * t)  #  10Hz 30Hz  

#       
frequencies = np.fft.fftfreq(len(t))
spectrum = fft(signal)
amplitude = np.abs(spectrum)

#    
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

#       .
ax1.plot(t, signal)
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Amplitude')
ax1.set_title('Time Domain Signal')

#       .
ax2.plot(frequencies, amplitude)
ax2.set_xlim(-50, 50)  #    
ax2.set_xlabel('Frequency (Hz)')
ax2.set_ylabel('Amplitude')
ax2.set_title('Frequency Domain Signal (Spectrum)')

plt.tight_layout()
plt.show()
```
