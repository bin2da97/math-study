## 크롬 드라이버 실행
크롬 드라이버는 컴퓨터가 크롬 웹 브라우져를 다룰 수 있도록 해주는 프로그램


```python
driver = webdriver.Chrome()
driver.get(url="https://news.naver.com/")
html=requests.get(load_url,headers={'User-Agent':'Mozilla/5.0'})
soup=BeautifulSoup(html.content,"html.parser")
f_url=soup.find_all("a")
news_url=[i.get('href') for i in f_url if i.get("class")==["Nitem_link"]]
result={}
for i in range(1,14):
    try:
        search_button=driver.find_element(By.XPATH , f'/html/body/section/header/div[2]/div/div/div[1]/div/div/ul/li[{i}]/a')
        search_button.click()
        load_url=news_url[i-1]
        print('url',load_url)
        #여러번 크롤링하면 오류발생하여 headers 추가 
        html=requests.get(load_url,headers={'User-Agent':'Mozilla/5.0'})
        soup=BeautifulSoup(html.content,"html.parser")
        n_url=soup.find_all("img")
        result[i]=result.get(i,[])+[i.get("src") for i in n_url]
        print('반복횟수:' ,i , 'result: ',result)
    except:
        print("예외 발생 ")
        pass
    time.sleep(3)
    driver.get(url="https://news.naver.com/")
```
